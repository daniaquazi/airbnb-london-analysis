# London Airbnb Analysis — SQL Analysis

This notebook loads the cleaned listings data into a SQLite database and uses SQL queries to answer questions about London Airbnb listings.

```python
# Import libraries
import pandas as pd
import sqlite3

print("Libraries loaded!")
```

## 1. Load Data and Create Database

We load the cleaned listings into a SQLite database so we can query it with SQL.

```python
# Load the cleaned listings data
listings = pd.read_csv('../data/listings_clean.csv')

print("Listings loaded!")
print("Rows:", len(listings))
```

```python
# Create a SQLite database
conn = sqlite3.connect(':memory:')

# Load listings into the database as a table
listings.to_sql('listings', conn, if_exists='replace', index=False)

print("Database ready!")
```

## 2. Simple Queries

Starting with basic SQL to understand the data.

```python
# Select all columns from the first 5 rows
query = "SELECT * FROM listings LIMIT 5"
result = pd.read_sql(query, conn)
print(result)
```

```python
# Select just the name, neighbourhood and price
query = "SELECT name, neighbourhood_cleansed, price FROM listings LIMIT 10"
result = pd.read_sql(query, conn)
print(result)
```

```python
# How many listings are there in total?
query = "SELECT COUNT(*) as total_listings FROM listings"
result = pd.read_sql(query, conn)
print(result)
```

```python
# What room types exist?
query = "SELECT DISTINCT room_type FROM listings"
result = pd.read_sql(query, conn)
print(result)
```

## 3. WHERE Queries

Filter the data to find specific listings.

```python
# Listings under £50 per night
query = "SELECT name, neighbourhood_cleansed, price FROM listings WHERE price < 50 LIMIT 10"
result = pd.read_sql(query, conn)
print(result)
```

```python
# Only entire home listings
query = "SELECT name, neighbourhood_cleansed, price FROM listings WHERE room_type = 'Entire home/apt' LIMIT 10"
result = pd.read_sql(query, conn)
print(result)
```

```python
# Only superhost listings
query = "SELECT name, neighbourhood_cleansed, price FROM listings WHERE host_is_superhost = 't' LIMIT 10"
result = pd.read_sql(query, conn)
print(result)
```

## 4. ORDER BY Queries

Sort the data to find the cheapest and most expensive listings.

```python
# 5 cheapest listings
query = "SELECT name, neighbourhood_cleansed, room_type, price FROM listings ORDER BY price ASC LIMIT 5"
result = pd.read_sql(query, conn)
print(result)
```

```python
# 5 most expensive listings
query = "SELECT name, neighbourhood_cleansed, room_type, price FROM listings ORDER BY price DESC LIMIT 5"
result = pd.read_sql(query, conn)
print(result)
```

## 5. GROUP BY Queries

Group the data to find averages and counts by category.

```python
# Average price per neighbourhood
query = """
SELECT neighbourhood_cleansed as neighbourhood,
       ROUND(AVG(price), 2) as avg_price,
       COUNT(*) as num_listings
FROM listings
GROUP BY neighbourhood_cleansed
ORDER BY avg_price DESC
LIMIT 10
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
# Count of listings by room type
query = """
SELECT room_type,
       COUNT(*) as num_listings,
       ROUND(AVG(price), 2) as avg_price
FROM listings
GROUP BY room_type
ORDER BY num_listings DESC
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
# Superhost vs regular host
query = """
SELECT host_is_superhost,
       COUNT(*) as num_listings,
       ROUND(AVG(price), 2) as avg_price,
       ROUND(AVG(review_scores_rating), 2) as avg_rating
FROM listings
GROUP BY host_is_superhost
"""
result = pd.read_sql(query, conn)
print(result)
```

## 6. Summary

Key findings from the SQL analysis based on 60,951 London Airbnb listings:

**Price by neighbourhood:**
- Most expensive borough: **City of London** at £276/night average (408 listings)
- Second most expensive: **Kensington and Chelsea** at £269/night (4,570 listings)
- **Westminster** third at £263/night with the most central listings (8,116)
- **Tower Hamlets** and **Southwark** are more affordable at ~£151/night

**Room types:**
- **Entire home/apt** dominates with 41,596 listings at an average of £215/night
- **Private rooms** are the budget option with 19,117 listings at £80/night
- **Hotel rooms** are rare (54 listings) but priced similarly to entire homes at £208/night
- **Shared rooms** are cheapest at £59/night but very uncommon (184 listings)

**Superhost vs regular host:**
- There are 14,319 superhosts vs 45,308 regular hosts
- Superhosts charge slightly more: **£185** vs **£168** for regular hosts
- Superhosts get significantly better ratings: **4.85** vs **4.60**
- This suggests superhost status is worth it — better service commands a premium

**Price extremes:**
- Cheapest listing: £7/night (shared room in Waltham Forest)
- Most expensive listings: £999/night (entire homes in Westminster and Tower Hamlets)
